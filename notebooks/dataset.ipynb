```python
import torch

def random_row_dropout_with_mask(tensor, min_ratio=0.15, max_ratio=0.30):
    """
    将 2-D torch 张量每行随机置零 15%-30% 的元素，并返回置零掩码。

    参数
    ----
    tensor : torch.Tensor
        形状 (m, n) 的输入张量。
    min_ratio : float
        每行最小置零比例。
    max_ratio : float
        每行最大置零比例。

    返回
    ----
    modified_tensor : torch.Tensor
        已就地修改的张量（与输入共享内存）。
    zero_mask : torch.Tensor
        形状 (m, n) 的 bool 张量，True 表示对应位置被置零。
    """
    assert tensor.dim() == 2, "输入必须是 2-D 张量"
    m, n = tensor.shape

    # 1) 为每行独立采样置零比例
    ratios = torch.empty(m, device=tensor.device).uniform_(min_ratio, max_ratio)
    # 2) 计算每行需置零的元素个数（至少 1 个，至多 n-1 个）
    num_zeros = torch.clamp((ratios * n).round().long(), max=n - 1)

    # 3) 构建掩码
    zero_mask = torch.zeros(m, n, dtype=torch.bool, device=tensor.device)
    for i in range(m):
        if num_zeros[i] > 0:
            idx = torch.randperm(n, device=tensor.device)[:num_zeros[i]]
            zero_mask[i, idx] = True

    # 4) 应用掩码
    tensor.masked_fill_(zero_mask, 0)

    return tensor, zero_mask
```

```python
elist = []
edict = {}
for e in e_include:
    edict[e] = 0
smiles_contains = []
for s in smiles:
    es = count_element(s)

    if np.all([e in e_include for e in es]):
        elist+=es
        for e in es:
            if e in e_include:
                edict[e]+=1
        smiles_contains.append(s)
```
